# Actividad: Mi empresa

La siguiente actividad pone en practica tus conocimientos alcanzados en la manipulacion de una base de datos sencilla de Access, crear tablas y formularios.

### Obtención del sello (valor  10 puntos)

- Nombre de la actividad: Mi empresa
- Ambito: Diseño de tablas, campos y organización de la información
- Competencia: 2
- Objetivo: Crear una base de datos correctamente acompañada de sus tablas con los campos correspondientes y formularios para organizar información.

## Desarrollo

### Primera parte

- En Google busca drawIO y realiza el diseño de las siguientes tablas como hemos aprendido en clases anteriores:

- Aplica color a las tablas uno para el nombre y otro para las demas filas

### Segunda parte:  Creacion de tablas y bases de datos en Access

1. Abre Access y crea una base de datos en blanco
2. Asiganle como nombre: ```miEmpresa``` (sin espacios)
3. Ve al menu crear y selecciona ```tabla```.
4. Comprueba que estas ubicado en ```inicio```
5. Haz clic en ```ver -> Vista diseño```
6. Crea la primera tabla teniendo en cuenta el diseño que realizaste en DrawIO
7. Repite el proceso con todas las tablas

### Tercera parte: creacion de los formularios

1. Verifica que haz creado todas las tablas
2. Ve al menu ```Crear``` y selecciona ```Asistente para formularios```
3. En la lista desplegable selcciona la tabla ```Productos```
4. Pasa al lado derecho con la flecha todos los campos __menos ```id_producto```__
5. Selecciona ```justificado```
6. Asigna como nombre al formulario ```formulario productos```.
7. Repite el proceso con las tablas ```cliente``` y ```empleado```.

### Cuarta parte: agregar los registros 

1. Ubicate en el ```formulario productos```
2. Agrega 10 productos que te gustaria vender, llenando toda la informacion.
3. Ahora ve al ```formulario cliente```
4. Agrega 5 nombres de personas que conozcas llenando los datos solicitados
5. Finalmente dirigete al ```formulario empleados``` y agrega 10 nombres de amigos, familiares o personas inventadas llenando la informacion.

Listo, guarda los cambios y llama al profesor para recibir el sello con tu nota.
