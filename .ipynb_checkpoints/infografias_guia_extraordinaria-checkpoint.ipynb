# Infografías

Una infografía es una representación visual de información que combina texto, imágenes, gráficos y diagramas para comunicar un mensaje de manera clara, atractiva y concisa. Su propósito es facilitar la comprensión de datos o conceptos complejos a través de elementos visuales bien estructurados.

Se utilizan en diversos ámbitos, como educación, periodismo, marketing y redes sociales, y suelen incluir elementos como:
- Títulos y subtítulos para organizar la información.
- Iconos y gráficos para ilustrar ideas clave.
- Colores y tipografías llamativas para captar la atención.
- Diagramas, mapas o tablas para presentar datos de forma visual.

Su principal ventaja es que permite absorber información rápidamente sin necesidad de leer extensos bloques de texto

## Que elementos componen una infografía

Una infografía debe integrar los siguientes elementos clave para ser efectiva y visualmente atractiva:

1. Título llamativo

- Debe ser corto, claro y reflejar el tema central de la infografía.

2. Subtítulos o secciones

- Organizan la información en partes comprensibles.
- Ayudan a guiar al lector a través del contenido.

3. Texto breve y claro

- Debe ser conciso y fácil de leer.
- Se usa para explicar los datos o conceptos esenciales.

4. Imágenes e íconos

- Refuerzan el mensaje visualmente.
- Hacen que la información sea más atractiva y comprensible.

5. Gráficos y diagramas

- Se utilizan para representar datos de forma visual (barras, líneas, pastel, etc.).
- Facilitan la comparación de información.

6. Colores y tipografías adecuadas

- Deben ser armoniosos y mantener una jerarquía visual clara.
- No deben saturar la infografía ni dificultar la lectura.

7. Fuente de información

- Si la infografía usa datos o referencias externas, debe incluir la fuente al final.

## Recursos para crear infografías

A continuación se presentan algunos recursos que puedes emplear para crear infografías

### Draw.io

- No requiere registro
- Diseño simple

[Enlace al sitio](https://app.diagrams.net)

> Las infografías se encuentran en la sección: 
>
> Archivo > Nuevo > Otros

### Piktochart

- Requiere una cuenta de Google
- Interfaz completa con cientos de ejemplos
- Especializada en infografías

[Enlace al sitio](https://piktochart.com//)

### Canva

- Requiere registro
- Interfaz completa
- Es un sitio web de propósito general

> Las infografías se encuentran en la seccion de "Marketing"

[Enlace al sitio](https://www.canva.com/)

## Actividad

1. Investiga sobre un tema de tu interés
2. Busca imagenes e iconos relacionados con el tema
3. Selecciona alguno de los recursos vistos
4. Crea tu infografia sobre el tema que investigaste, el titulo debe ser el nombre del tema, agrega textos muy cortos, imagenes muy representativas sobre el tema e iconos.
5. La infografía debe dar un resumen o panorama general del tema.

### Obtención del sello

Escriba en su cuaderno:

__Título de la actividad__: Creación de infografías

__Ambito Conceptual__:Representación de simbolos y sintesis de conceptos de forma descriptiva y gráfica.

__Competencia__: Diseñar infografías que transmitan de forma clara y gráfica los elementos claves de un tema particular para desarrollar la creatividad y la expresión simbólica escrita.

__Objetivo__: Crear una infografía haciendo uso de alguno de los recursos propuestos para transmitir la idea sobre un tema de manera concreta y gráfica.
